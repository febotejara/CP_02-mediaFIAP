**Desafio: Cálculo da média da FIAP**

```python
# cálculo de média dos checkpoints
# 3 CPs obrigatórias -----> input
# Estrutura de seleçãi IF, ELIF E ELSE
# Eliminar a menor nota
# Calcular média das duas maiores notas (MCP_1SEM)
```

```python
CP1, CP2, CP3 = input('Inisira as notas dos checkpoints: ').split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)
```

```python
CP12, CP22, CP32 = input('Inisira as notas dos checkpoints: ').split()
CP12 = float(CP12)
CP22 = float(CP22)
CP32 = float(CP32)
```

```python
if (CP1 > CP2):
  if (CP2 < CP3):
    MCP_1SEM = (CP1 + CP3) / 2
    print (f'CP2 descartada, média CPs: {MCP_1SEM:.1f}.')

elif (CP1 < CP2):
  if (CP1 < CP3):
    MCP_1SEM = (CP2 + CP3) / 2
    print (f'CP1 descartada, média CPs: {MCP_1SEM:.1f}.')

if (CP3 < CP2):
  if (CP3 < CP1):
    MCP_1SEM = (CP1 + CP2) / 2
    print (f'CP3 descartada, , média CPs: {MCP_1SEM:.1f}')
```

```python
if (CP12 > CP22):
  if (CP22 < CP32):
    MCP_2SEM = (CP12 + CP32) / 2
    print (f'CP2 descartada, média CPs: {MCP_2SEM:.1f}.')

elif (CP12 < CP22):
  if (CP12 < CP32):
    MCP_2SEM = (CP22 + CP32) / 2
    print (f'CP1 descartada, média CPs: {MCP_2SEM:.1f}.')

if (CP32 < CP22):
  if (CP32 < CP12):
    MCP_2SEM = (CP12 + CP22) / 2
    print (f'CP3 descartada, , média CPs: {MCP_2SEM:.1f}')
```

```python
# Cálculos de notas de SPINTs (SP1 e SP2)
# Média das duas notas
SP1, SP2 = input('Inisira as notas das Sprints: ').split()
SP1 = float(SP1)
SP2 = float(SP2)
```

```python
SP12, SP22 = input('Inisira as notas das Sprints: ').split()
SP12 = float(SP12)
SP22 = float(SP22)
```

```python
# Média SPs
MSP_1SEM = (SP1 + SP2) / 2
print (f'Média Sprints: {MSP_1SEM:.1f}.')
```

```python
MSP_2SEM = (SP12 + SP22) / 2
print (f'Média Sprints: {MSP_2SEM:.1f}.')
```

```python
# Cálculo de nota de Global Solution (GS)
# Inserir a nota.
# Caso não haja nota de GS, o aluno deverá inserir sua nota SUB

GS1 = input('Inisira a nota da GS (caso tenha feito SUB deixe vazio): ')

if (GS1 != ""):
  GS1 = float(GS1)
  print (f'Nota GS1 = {GS1}')

else:
  GS1 = float(input('Inisira a nota da GS (caso tenha feito SUB deixe vazio): '))
  print (f'Nota SUB = {GS1}')
```

```python
GS2 = input('Inisira a nota da GS (caso tenha feito SUB deixe vazio): ')

if (GS2 != ""):
  GS2 = float(GS2)
  print (f'Nota GS2 = {GS2}')

else:
  GS2 = float(input('Inisira a nota da GS (caso tenha feito SUB deixe vazio): '))
  print (f'Nota SUB = {GS2}')
```

```python
# Cálculo da média semestral 1 (MS1)
MS1 = 0.6*GS1 + 0.2*MSP_1SEM + 0.2*MCP_1SEM
print (f'Média Semestral: {MS1:.1f}.')

# repetir para 2 semestre
MS2 = GS2*0.6 + MSP_2SEM*0.2 + MCP_2SEM*0.2
print (f'Média Semestral: {MS2:.1f}.')
```

```python
# média final fiap
MF = 0.4*MS1 + 0.6*MS2
```

```python
# apresentar se o aluno está aporvado ou de exame >= 6.0
# se estiver de exame, inserir a nota do exame >= 4.0 e < 6.0
# repeovado < 4.0
if (MF >= 6.0):
  print (f'Aluno aprovado: {MF:.1f}')

else:
  if (MF >=4.0 < 6.0):
    print (f'Aluno de exame: {MF:.1f}')

    #se stiver de exame inserir a nota do exame
    EX = float(input('Inisira a nota do exame: '))

    MF2 = (EX-MF) / 2
    print (f'Sua nova média após exame é: {EX:.1f}')


  if (MF >= 6.0):
    print (f'Aluno aprovado no exame: {MF:.1f}')


  else:
    print (f'Aluno reprovado no exame!')
```
