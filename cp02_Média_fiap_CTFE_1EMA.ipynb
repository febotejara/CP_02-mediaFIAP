**Desafio: Cálculo da média da FIAP**

```python
# cálculo de média dos checkpoints
# 3 CPs obrigatórias -----> input
# Estrutura de seleçãi IF, ELIF E ELSE
# Eliminar a menor nota
# Calcular média das duas maiores notas (MCP_1SEM)
```

```python
CP1, CP2, CP3 = input('Inisira as notas dos checkpoints: ').split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)
```

```python
if (CP1 > CP2):
  if (CP2 < CP3):
    MCP_1SEM = (CP1 + CP3) / 2
    print (f'CP2 descartada, média CPs: {MCP_1SEM:.1f}.')

elif (CP1 < CP2):
  if (CP1 < CP3):
    MCP_1SEM = (CP2 + CP3) / 2
    print (f'CP1 descartada, média CPs: {MCP_1SEM:.1f}.')

if (CP3 < CP2):
  if (CP3 < CP1):
    MCP_1SEM = (CP1 + CP2) / 2
    print (f'CP3 descartada, , média CPs: {MCP_1SEM:.1f}')
```

```python
# Cálculos de notas de SPINTs (SP1 e SP2)
# Média das duas notas
SP1, SP2 = input('Inisira as notas das Sprints: ').split()
SP1 = float(SP1)
SP2 = float(SP2)
```

```python
# Média SPs
MSP_1SEM = (SP1 + SP2) / 2
print (f'Média Sprints: {MSP_1SEM:.1f}.')
```

```python
# Cálculo da média semestral 1 (MS1)
MS1 = (GS1 * 0.6 + MSP_1SEM * 0.2 + MCP_1SEM * 0.2)
print (f'Média Semestral: {MS1:.1f}.')

# repetir para 2 semestre
```
